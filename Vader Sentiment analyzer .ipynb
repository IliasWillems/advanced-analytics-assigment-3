```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Read in the CSV file
df = pd.read_csv("C:/Users/tibo_/Downloads/spark/spark/notebooks/data/review_data.csv")

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Define a function to apply VADER sentiment analysis to a text string
def get_sentiment(text):
    text_str = str(text) # convert to string
    sentiment = analyzer.polarity_scores(text_str)
    if sentiment['compound'] > 0:
        return 'Positive'
    else:
        return 'Negative'

# Apply the get_sentiment function to the review_text column and create a new column with the results
df['sentiment'] = df['review_text'].apply(get_sentiment)

# Print the resulting DataFrame
print(df.head(40))

```
